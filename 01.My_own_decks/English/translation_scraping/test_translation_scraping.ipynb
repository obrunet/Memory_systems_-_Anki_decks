First let's retrieve a list of the most frequently used english words here:  
https://www.ef.com/wwen/english-resources/english-vocabulary/top-3000-words/


```python
filename = "./3000 most common words in English _ EF _ Global Site.txt"

with open(filename) as file:
    words = file.readlines()
    words = [word.rstrip() for word in words]
    
words[300:310]
```

Then let's scrape translations for each word from:  

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from requests.exceptions import HTTPError

#from random import randint
#from time import sleep
```

```python
url = "https://dictionary.cambridge.org/fr/dictionnaire/anglais-francais/tired"

try:
    headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:53.0) Gecko/20100101 Firefox/53.0"} # UserAgent().random}
    req_response = requests.get(url, headers=headers)
    # If the response was successful, no Exception will be raised
    req_response.raise_for_status()
except HTTPError as http_err:
    print(f'HTTP error occurred: {http_err}')
except Exception as err:
    print(f'Other error occurred: {err}')
else:
    print('Request success!')
```

```python
s = BeautifulSoup(req_response.content, 'html.parser')
s.prettify()
```

Finally, let's do it for all the words in a list

```python
url = "https://dictionary.cambridge.org/fr/dictionnaire/anglais-francais/"
word_list, translations = words, []


for w in word_list:
    try:
        headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:53.0) Gecko/20100101 Firefox/53.0"} # UserAgent().random}
        req_response = requests.get(url + w, headers=headers)
        req_response.raise_for_status()
    except:
        result = "Nan"
    else:
        s = BeautifulSoup(req_response.content, 'html.parser')
        result = ", ".join(set([i.get_text().split(" ")[0] for i in s.find_all("span", class_="trans dtrans", lang="fr")]))
    translations.append(result)
    

len(word_list) == len(translations)
```

```python
df = pd.DataFrame(list(zip(word_list, translations)), columns =['word_list', 'translations'])
df.head(50)
```

```python
df.to_csv('translation_eng_fr.csv', index=False)
```
