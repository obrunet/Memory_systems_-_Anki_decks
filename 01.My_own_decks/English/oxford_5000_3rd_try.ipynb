```python
import requests
from bs4 import BeautifulSoup
from requests.exceptions import HTTPError

url = "https://dictionary.cambridge.org/fr/dictionnaire/anglais-francais/"
word_list, translations = list(df['word']), []


for w in word_list:
    try:
        headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:53.0) Gecko/20100101 Firefox/53.0"} # UserAgent().random}
        req_response = requests.get(url + w, headers=headers)
        req_response.raise_for_status()
        
        
    except HTTPError as http_err:
        # print(f'HTTP error occurred: {http_err}')
        result = "Nan"
    except Exception as err:
        # print(f'Other error occurred: {err}')
        result = "Nan"
    else:
        # print('Request success!')        
        s = BeautifulSoup(req_response.content, 'html.parser')
        result = ", ".join(set([i.get_text().split(" ")[0] for i in s.find_all("span", class_="trans dtrans", lang="fr")]))
    translations.append(result)
    

tr = pd.DataFrame(list(zip(word_list, translations)), columns =['word_list', 'translations'])
tr.to_csv('translation_eng_fr.csv', index=False)
tr.head()
```
