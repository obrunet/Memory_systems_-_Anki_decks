## download sounds

```python
df['source_and_online_sound'][1]
```

```python
df['source_and_online_sound'][1].split(" ")[-1]
```

```python
df['sound_url'] = df['source_and_online_sound'].apply(lambda x: x.split(" ")[-1])
df['sound_url'][1]
```

```python
df['sound_file'] = df['sound_url'].apply(lambda x: x.split("/")[-1])
df['sound_file'][1]
```
